```python
##from sentence_transformers import SentenceTransformer, util
from sentence_transformers import SentenceTransformer, util
from collections import Counter

# Load a pre-trained sentence transformer model
##model = SentenceTransformer('paraphrase-MiniLM-L6-v2')  Lets try below one
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Define the target categories
categories = ["Processing Power", "Absorption", "Accountability", "Integrity", "Agility and Adaptability",
              "Stress-Based Leadership", "Becoming a High Performance Coach", "Bold Strategy", "Courage", "Building Trust",
              "Building a High Performance Team", "Change Leadership", "Inspirational Leadership", "Communication",
              "Customer Centricity", "Coaching and Talent Development", "Collaboration", "Horizontal Leadership",
              "Continuous Learning", "Creativity", "Culturally Astute", "Curiosity", "Data-Driven", "Decision Making",
              "Decisive Leadership", "Delegation", "Consensus Leadership", "Inclusive Leadership", "Driving Accountability",
              "Emotional Consistency", "Emotional Intelligence", "Resilience", "Engagement", "Enterprise Leadership",
              "Entrepreneurial Mindset", "Operational Excellence", "Executive Presence", "Experiential Leadership",
              "Extraversion", "Humility", "Influence", "Innovation", "Intelligence", "Judgment", "Triangulation",
              "Distributed Leadership", "Listening", "Learning Orientation", "Developing People", "Navigating Ambiguity",
              "Negotiation and Deal-Making", "Openness to Experience", "Operational Excellence", "Operational Leadership",
              "Directive Leadership", "Perseverance", "Personal Leadership", "Perspective", "Prioritization",
              "Process and Systems Thinking", "Influential", "People-Focused Leadership", "War-Time Leadership",
              "Energy Management", "Servant Leadership", "Social Intelligence", "Stakeholder Management", "Strategic Thinking",
              "Strategy", "Talent Management", "Process-Oriented Leadership", "Team Leadership", "Teamwork",
              "Tensioning the Organization", "Thought Leadership", "Performance Leadership", "Transformational Leadership",
              "Transformative Change", "Visibility", "Talent Acquisition", "Talent / Team Building", "One-to-Many Communication",
              "Outside-In Perspective", "Personal Ambition", "Enterprise Ambition"]

def assign_categories(responses, categories):
    assigned_categories = {}
    category_count = Counter()

    for person, sentences in responses.items():
        sentence_embeddings = model.encode(sentences)
        category_embeddings = model.encode(categories)

        used_indices = set()
        person_results = {}

        for i, sentence_embedding in enumerate(sentence_embeddings):
            best_match = None
            best_score = -1
            best_index = -1

            for j, category_embedding in enumerate(category_embeddings):
                if j in used_indices:
                    continue

                # Compute cosine similarity
                similarity = util.cos_sim(sentence_embedding, category_embedding).item()

                if similarity > best_score:
                    best_score = similarity
                    best_match = categories[j]
                    best_index = j

            # Assign the best match and mark the category as used
            person_results[sentences[i]] = best_match
            used_indices.add(best_index)
            category_count[best_match] += 1

        assigned_categories[person] = person_results

    return assigned_categories, category_count

# Example input for Peers
peer_inputs = {
    "Peer1": [
        "Marty is good at talking to people and making sure they feel heard. He is open and listens to different ideas, even if he doesn’t agree.",
        "He is great at building relationships and working with different teams. He also knows a lot about how the business runs, not just his own area."
    ],
    "Peer2": [
        "Marty is friendly and always asks how people are doing. He really cares about his team and helps when he can.",
        "He is smart and fair, always listening to different opinions before making a decision. He also knows a lot about different parts of the company."
    ],
    "Peer3": [
        "Marty is really good at explaining things in a way that makes sense. He helps make big changes without making people feel stressed.",
        "He is great at working with different teams and making sure everyone is on the same page. His planning skills help get things done the right way."
    ],

    "Peer4": [
        "Marty listens to people and cares about what they think. He works hard to make sure changes are fair and helpful.",
        "He helps his team think about the big picture and how things will work in the future. He also trusts people to do their jobs without checking on them too much."
    ]
}


# Example input for Direct Reports
direct_report_inputs = {
    "DirectReport1": [
        "Marty is really nice and talks to everyone, making sure people feel heard. He also explains things in a way that makes sense.",
        "He is great at getting teams to work together and making sure everyone knows what’s going on. He also visits different locations and checks in with teams."
    ],
    "DirectReport2": [
        "Marty is a great leader who listens and cares about his team. He takes time to explain why things are changing.",
        "He helps make work easier by fixing slow processes and getting rid of useless meetings. He also makes sure people feel valued."
    ],
    "DirectReport3": [
        "Marty is smart and knows how to get people to work together. He is friendly and listens well, but sometimes he seems too busy to check in.",
        "He is great at making teams stronger and helping them solve problems. He also has a good way of keeping people focused on the goal."
    ],
    "DirectReport4": [
        "Marty is good at bringing people together and making work run smoothly. He works hard and makes sure things get done.",
        "He is honest and makes sure people know what’s expected of them. He also speaks up when something isn’t right."
    ],
    "DirectReport5": [
        "Marty listens and is open to honest feedback, but sometimes people have to reach out to him first. He is good at solving problems but needs to check in with his team more often.",
        "He trusts his leaders and helps them grow by giving them space to make decisions. He also makes sure frontline employees feel valued by stopping in and talking to them."
    ]
}

# Process Peer feedback
peer_results, peer_category_count = assign_categories(peer_inputs, categories)

# Process Direct Report feedback
direct_report_results, direct_report_category_count = assign_categories(direct_report_inputs, categories)

# Combine both counts for an overall ranking
overall_category_count = peer_category_count + direct_report_category_count

# Display individual peer results
print("\nPeer Feedback:")
for peer, responses in peer_results.items():
    print(f"\n{peer}'s Responses:")
    for sentence, category in responses.items():
        print(f"  Input: {sentence}")
        print(f"  Output: {category}\n")

# Display individual direct report results
print("\nDirect Report Feedback:")
for report, responses in direct_report_results.items():
    print(f"\n{report}'s Responses:")
    for sentence, category in responses.items():
        print(f"  Input: {sentence}")
        print(f"  Output: {category}\n")

# Display Peer Top 3 Strengths
print("\nPeer Top 3 Categories:")
for category, count in peer_category_count.most_common(3):
    print(f"  {category}: {count} times")

# Display Direct Report Top 3 Strengths
print("\nDirect Report Top 3 Categories:")
for category, count in direct_report_category_count.most_common(3):
    print(f"  {category}: {count} times")

# Display Overall Top 3 Categories
print("\nOverall Top 3 Categories:")
for category, count in overall_category_count.most_common(3):
    print(f"  {category}: {count} times")
```
